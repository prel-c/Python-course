Index - спобос сослатсья (reference) на данные в Series или в DataFrame

```python
index = pd.Index([2, 5, 3, 6, 71])
print(index)
print(type(index))
print(index[1])
print(index[::2])
```

```python
#index[1] = 123 - индексы менять нельзя
```

```python
index1 = pd.Index([2, 5, 3, 6, 71])
index2 = pd.Index([1, 2, 51, 71, 41])
print(index1.intersection(index2))
print(index1.union(index2))
print(index1.symmetric_difference(index2))
```

```python
dict1 = {
    'A':10,
    'B':20,
    'C':30,
    'D':40,
    'E':50,
}
data_dict1 = pd.Series(dict1)

dict2 = {
    'A':11,
    'B':21,
    'C':31,
    'D':41,
    'E':51,
}
data_dict2 = pd.Series(dict2)

df = pd.DataFrame({"dict_01":data_dict1, "dict_02":data_dict2})
print(df)
```

```python
print(df["dict_01"])
print(df.dict_01)
```

```python
df["new"] = df["dict_01"]
print(df)
```

```python
df["new"] = df["dict_01"] / df["dict_02"]
print(df)
```

```python
print(df.values)
print(df["dict_01"]) # столбец
print(df.values[0]) # строка
```

не смотря на то, что numpy ничего не знает про pandas, операции делать можно и индексы сохраняются

```python
dict1 = {
    'A':10,
    'B':20,
    'C':30,
    'D':40,
    'E':50,
    'F':60,
}
data_dict1 = pd.Series(dict1)

dict2 = {
    'A':11,
    'B':21,
    'C':31,
    'D':41,
    'E':51,
    'H':71
}
data_dict2 = pd.Series(dict2)


df = pd.DataFrame({"dict_01":data_dict1, "dict_02":data_dict2})
print(df)

print(data_dict1 + data_dict2)
print(np.add(data_dict1, data_dict2))
print(data_dict1.add(data_dict2, fill_value = 5))
```

```python
df1 = pd.DataFrame({"dict_10":data_dict1, "dict_02":data_dict2})
print(df1)

print(df + df1)

print(df.add(df1, fill_value=555))
print(df.add(df1, fill_value=df.values.sum()))
```
