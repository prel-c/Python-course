```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-10, 10, 50)
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

#ax.scatter3D(X, Y, Z, c=Z)
#ax.plot_wireframe(X, Y, Z)

ax.plot_surface(X, Y, Z, cmap='viridis')


plt.show()

```

```python
angle = np.linspace(0, 2 * np.pi, 50)
r = np.linspace(0, 6, 30) 
R, Angle = np.meshgrid(r, angle)

ax = plt.axes(projection='3d')

X = R * np.sin(Angle)
Y = R * np.cos(Angle)

Z = f(X, Y)

ax.plot_surface(X, Y, Z, cmap='viridis')

plt.show()
```

```python
angle = np.linspace(0,  1.5 * np.pi, 50)
r = np.linspace(0, 6, 30) 
R, Angle = np.meshgrid(r, angle)

ax = plt.axes(projection='3d')

X = R * np.sin(Angle)
Y = R * np.cos(Angle)

Z = f(X, Y)

ax.plot_surface(X, Y, Z, cmap='viridis')


plt.show()
```

```python
angle = np.linspace(0,  1.5 * np.pi, 1000)
r = np.linspace(0, 6, 1000) 
R, Angle = np.meshgrid(r, angle)

ax = plt.axes(projection='3d')

X = R * np.sin(Angle)
Y = R * np.cos(Angle)
Z = f(X, Y)

x = r * np.sin(angle)
y = r * np.cos(angle)
z = f(x, y)


ax.plot_trisurf(x, y, z, cmap='viridis')

plt.show()
```

```python
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

```python
cars = pd.read_csv("D:/Obsidian/base/-Python/Python libraries/assets/cars.csv")
print(cars.head())
```

```python
#парная карта
sns.pairplot(data=cars, hue="transmission")
```

```python
# тепловая карта

cars_corr = cars[['year', "selling_price", "seats", "mileage"]]

sns.heatmap(cars_corr.corr(), cmap="viridis", annot=True)

plt.show()
```

```python
# диаграмма рассеивания

sns.scatterplot(data=cars, x="year", y='selling_price')

plt.show()
```

```python
sns.scatterplot(data=cars, x="seats", y='mileage', hue='fuel')

plt.show()
```

```python
# диаграмма рассеивания + линейная регрессия

sns.regplot(data=cars, x="seats", y='mileage')

plt.show()
```

```python
sns.relplot(data=cars, x="seats", y='mileage', kind='scatter',
            col='fuel', col_wrap=2, hue='transmission')

plt.show()
```

```python
sns.relplot(data=cars, x="seats", y='mileage', kind='line',
            col='fuel', col_wrap=2, hue='transmission')

plt.show()
```

```python
sns.lmplot(data=cars, x="seats", y='mileage',
            col='transmission', col_wrap=2, hue='fuel')

plt.show()
```

```python
# линейный график

sns.lineplot(data=cars, x="seats", y='mileage')

plt.show()
```
