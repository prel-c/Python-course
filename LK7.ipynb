работа с временными рядами
1) 1 января 2026 года 00:00
2) период, март 2025, 24 часа
3) дельта: продолжительность 15 минут

```python
from datetime import datetime
import dateutil
```

```python
d = datetime(year=2026, month=3, day=4)
print(d, type(d))
```

```python
d = dateutil.parser.parse("4th of March, 2026")
print(d.strftime("%A"))
```

```python
t = np.array(12, dtype=np.timedelta64)
print(t)
```

```python
d = np.datetime64("2026-03-04", "D")
print(d)
print(d + 1)
```

```python
t = np.timedelta64(1, "D")
print(t, t + 1)
```

```python
d = pd.to_datetime("4th of March, 2026")
print(d, type(d))
```

```python
d2 = d + pd.to_timedelta(np.arange(12), "D")
print(d2)
```

```python
index = d2
data = pd.Series(np.arange(12), index=index)
print(data["2026-03-06":"2026-03-10"])
print(data["2026-03"])
```

```python
ds = pd.to_datetime(["2026-03-04", "2026-03-05"])

p = ds.to_period("D") 
print(p, type(p))
```

```python
delt = ds[1] - ds[0]
print(delt, type(delt))
idelt = ds - ds[0]
print(idelt, type(idelt))
```
