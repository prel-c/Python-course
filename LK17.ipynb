```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
print (iris.head())
```

```python
species_int = []
for row in iris.values:
    match row[4]:
        case "setosa":
            species_int.append(1)
        case "versicolor":
            species_int.append(2)
        case "virginica":
            species_int.append(3)
```

```python
species_int_df = pd.DataFrame(species_int)
print(species_int_df.head())
```

```python
data = iris[["sepal_length", "petal_length"]]
data["species"] = species_int

print(data.head())
print(data.shape)

data_df = data[(data["species"] == 1) | (data["species"] == 2)]
print(data_df.shape)
```

```python
data_of_setosa = data[data["species"] == 1]
data_of_versicolor = data[data["species"] == 2]
data_of_virginica = data[data["species"] == 3]

plt.scatter(data_of_setosa["sepal_length"], data_of_setosa["petal_length"])
plt.scatter(data_of_versicolor["sepal_length"], data_of_versicolor["petal_length"])

X = data_df[["sepal_length", "petal_length"]]
y = data_df["species"]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
x1_p = np.linspace(min(data_df["sepal_length"]), max(data_df["sepal_length"]), 100)
x2_p = np.linspace(min(data_df["petal_length"]), max(data_df["petal_length"]), 100)

X1_p, X2_p = np.meshgrid(x1_p, x2_p)

print(X1_p.shape)

X_p = pd.DataFrame(np.vstack([X1_p.ravel(), X2_p.ravel()]).T, columns=["sepal_length", "petal_length"])

print (X_p.head())

y_p = model.predict(X_p)

print (y_p)

plt.scatter(data_of_setosa["sepal_length"], data_of_setosa["petal_length"])
plt.scatter(data_of_versicolor["sepal_length"], data_of_versicolor["petal_length"])

plt.contourf(
    X1_p,
    X2_p,
    y_p.reshape(X1_p.shape),
    alpha=0.3,
    levels=[0, 1.5, 2.5]
)

plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
iris = sns.load_dataset('iris')
```

```python
data = iris[['sepal_length', 'petal_length', 'species']]
data_setosa = data[data['species'] == 'setosa'].drop(columns = ['species'])
```

```python
X = data_setosa['sepal_length']
Y = data_setosa['petal_length']
```

```python
pca = PCA(n_components=2)
pca.fit(data_setosa)
```

```python
plt.scatter(X, Y)

plt.scatter(pca.mean_[0], pca.mean_[1])
plt.plot(
    [pca.mean_[0], pca.mean_[0]+pca.components_[0][0]*np.sqrt(pca.explained_variance_[0])],
    [pca.mean_[1], pca.mean_[1]+pca.components_[0][1]*np.sqrt(pca.explained_variance_[0])],
)

plt.plot(
    [pca.mean_[0], pca.mean_[0]+pca.components_[1][0]*np.sqrt(pca.explained_variance_[1])],
    [pca.mean_[1], pca.mean_[1]+pca.components_[1][1]*np.sqrt(pca.explained_variance_[1])],
)
```

```python
pca1=PCA(n_components=1)
pca1.fit(data_setosa)
X_pca1=pca1.transform(data_setosa)
X_new=pca1.inverse_transform(X_pca1)

plt.scatter(X_new[:,0], X_new[:,1])
plt.tight_layout()
```
