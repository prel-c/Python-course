```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
x = np.arange(1, 21, 4)
y1 = np.array([1, 3, 4, 7, 12])
y2 = np.array([3, 2, 5, 12, 8])

plt.plot(x, y1, 'r--*', label='line1')
plt.plot(x, y2, 'g:>', label='line2')
plt.legend()
```

```python
x = np.arange(1, 21, 4)
xbig = np.arange(1, 41, 4)
y1 = np.array([1, 3, 4, 7, 12])
y2 = np.array([3, 2, 5, 12, 8])
ybig = np.array([3, 2, 4, 20, 34, 12, 24, 17, 10, 2])

plt.subplot(2, 1, 1)
plt.plot(xbig, ybig)

plt.subplot(2, 2, 3)
plt.plot(x, y1)

plt.subplot(2, 2, 4)
plt.plot(x, y2)

plt.plot()
```

```python
x = np.arange(-10, 11, 1)
y = np.array([i ** 2 for i in x])

fig, ax = plt.subplots()

ax.annotate('min', (0, 0), (0, 30), arrowprops={'facecolor': 'black', 'shrink': 0.1 })

plt.plot(x, y)
```

```python
x = np.random.randint(0, 255, (30, 30))
y = np.random.randint(0, 10, 100)

fig, ax = plt.subplots()

pcm = plt.pcolormesh(x)
axes = inset_axes(ax, width='5%', height='50%', loc='lower left', bbox_to_anchor=(1.05, 0., 1, 1), bbox_transform=ax.transAxes)

plt.colorbar(pcm, cax=axes)
```

```python
fig, ax = plt.subplots()

x = np.arange(0, 4*np.pi, np.pi/12)

y = np.array([np.sin(i) for i in x])

ax.plot(x, y)

ax.fill_between(x, y, color='r')
```

```python
fig, ax = plt.subplots()

x = np.arange(0, 4*np.pi, np.pi/120)

y = np.array([np.cos(i)+0.5 for i in x])

y = np.where(y >= 0.5, y, np.nan)

ax.plot(x, y)
ax.set_ylim(0, 1.6)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (12, 4))

x = np.arange(0, 13, 1)

ax[0].step(x, x, '-go')
ax[1].step(x, x, '-go', where='mid')
ax[2].step(x, x, '-go', where='post')
```

```python
fig, ax = plt.subplots()

x = np.arange(0, 13, 1)
y = np.array([i ** 3 for i in x])

ax.stackplot(x, y, y, y)
```

```python
fig, ax = plt.subplots()

x = np.array([10, 42, 32, 17, 80])
labels = ['pandas', 'numpy', 'pytorch', 'seaborn', 'matplotlib']
exp = [0, 0, 0, 0, 0.1]

ax.pie(x, labels=labels, explode=exp)
```
