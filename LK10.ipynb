```python
x = np.linspace(0, 10, 100)
ax = plt.gca()
ax.plot(x, np.sin(x), color="blue")
ax.plot(x, np.sin(x - 1), color="g", linestyle = "solid")
ax.plot(x, np.sin(x - 2), color="0.75", linestyle = "dashed") # оттенки серого
ax.plot(x, np.sin(x - 3), color="#FFDDAA", linestyle = "dashdot")
ax.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3), linestyle = "dotted")
ax.plot(x, np.sin(x - 5), color="salmon")
ax.plot(x, np.sin(x - 6), '--g')
plt.show()
```

Диаграмма рассеивания

```python
x = np.linspace(0, 10, 100)
print(x, type(x))
y = np.sin(x)
plt.plot(x, y, "--o", markersize=10, linewidth=4,
        markerfacecolor="red",
        markeredgecolor="blue",
        markeredgewidth=3,    color="black")
```

```python
rng = np.random.default_rng(1)
colors = rng.random(100)
sizes = 1000 * rng.random(100)
```

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.colorbar()
```

```python
from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)
print(iris.frame.head())
plt.scatter(
    iris.frame["sepal length (cm)"],
    iris.frame["sepal width (cm)"],
    s = 100 * iris.frame["petal length (cm)"],
    c = iris.frame["target"],
    alpha = 0.5
)
plt.show()
```

```python
rng = np.random.default_rng(1)
x = np.linspace(0, 10, 100)
dy = 0.5
y = np.sin(x) + dy*rng.random(100)
plt.plot(x, y, "-k")
plt.errorbar(x, y, yerr = dy, fmt=".r")
plt.fill_between(x, y - dy, y + dy, color = "0.75", alpha=0.7)
plt.show()
```

```python
# контурные графики
x = np.linspace(0, 6, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
print(X, X.shape)
print(Y, Y.shape)
```

```python
def f(f, y):
    return np.sin(x + 1) ** 4 + np.cos(y*x + 3)
Z = f(X, Y)
#print(Z.shape)
cnt = plt.contour(X, Y, Z, colors = "black")
plt.clabel(cnt, inline = True)
plt.imshow(
    Z, extent=[0,6, 0,5],
    origin = "lower",
    cmap = "RdGy",
    interpolation="gaussian",
    aspect="equal"
)
plt.colorbar()
plt.show()
```

```python
# Гистограммы
rng = np.random.default_rng(1)
data = rng.normal(size=100)
plt.hist(data, bins=30)
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)
p = dict(alpha=0.3, bins=40)
plt.hist(x1, **p)
plt.hist(x2, **p)
plt.hist(x3, **p)
plt.show()
print(np.histogram(x1, bins=40))
```

```python
mean = [0, 0]
cov = [[1,1], [1, 3]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x, y, bins = 40)
print(np.histogram2d(x, y, bins=2))
plt.show()
```
